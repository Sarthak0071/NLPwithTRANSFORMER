```python
from datasets import load_dataset, DatasetDict

# Load your dataset

# Split the dataset into training and testing subsets
split_dataset = dataset_dict['train'].train_test_split(test_size=0.2)

# Create a DatasetDict with the desired format
dataset = DatasetDict({
    "train": split_dataset["train"],
    "test": split_dataset["test"]
})

# Print the formatted dataset
print(dataset)
```

```python
from transformers import AutoTokenizer

tokenizer = AutoTokenizer.from_pretrained("distilbert-base-uncased")
```

```python
dataset['train'][14]
```

```python

def preprocess_function(examples):
    questions = [q.strip() for q in examples["question"]]
    answers = examples["answer"]
    inputs = tokenizer(
        questions,
        answers,
        max_length=384,
        truncation="only_second",
        padding="max_length",
        return_offsets_mapping=True,
    )

    offset_mapping = inputs.pop("offset_mapping")

    start_positions = []
    end_positions = []

    for i, offset in enumerate(offset_mapping):
        start_char = answers[i].find(answers[i])
        end_char = start_char + len(answers[i])
        start_positions.append(start_char)
        end_positions.append(end_char)

    inputs["start_positions"] = start_positions
    inputs["end_positions"] = end_positions
    return inputs
```

```python
tokenized_dataset = dataset.map(preprocess_function, batched=True, remove_columns=dataset["train"].column_names)
```

```python
from transformers import DefaultDataCollator

data_collator = DefaultDataCollator()
```

```python
import torch
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
device
```

```python
from transformers import AutoModelForQuestionAnswering, TrainingArguments, Trainer

model = AutoModelForQuestionAnswering.from_pretrained("distilbert-base-uncased").to(device)
```

```python
from huggingface_hub import notebook_login

notebook_login()
```

```python
training_args = TrainingArguments(

    output_dir="health_question",

    evaluation_strategy="epoch",

    learning_rate=2e-5,

    per_device_train_batch_size=16,

    per_device_eval_batch_size=16,

    num_train_epochs=2,

    weight_decay=0.01,

    push_to_hub=True,

)
```

```python
trainer = Trainer(

    model=model,

    args=training_args,

    train_dataset=tokenized_dataset["train"],

    eval_dataset=tokenized_dataset["test"],

    tokenizer=tokenizer,

    data_collator=data_collator,

)
```

```python
trainer.train()
```

```python
trainer.push_to_hub()
```
